```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
import matplotlib.pyplot as plt

# load your data into a numpy array, assuming the shape is (n_samples, n_features)

# specify the range of cluster numbers to evaluate
k_range = range(2, 100)

# initialize an empty list to store the average silhouette scores for each value of k
silhouette_scores = []

# loop through each value of k and compute the silhouette score for each object in the dataset
for k in k_range:
    # apply KMeans clustering with k clusters and fit the data
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    
    # compute the average silhouette score for all objects in the dataset
    silhouette_avg = silhouette_score(X, labels)
    silhouette_scores.append(silhouette_avg)

# plot the average silhouette score against the number of clusters
plt.plot(k_range, silhouette_scores, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Average Silhouette Score')
plt.title('Silhouette Analysis for KMeans Clustering')
plt.show()

# select the value of k that gives the highest silhouette score as the optimal number of clusters
optimal_k = k_range[np.argmax(silhouette_scores)]
print('Optimal number of clusters:', optimal_k)
```

## Problem 3
We adopted the strategy of Matrix Inverse by assuming the functional relation (1) to be linear:
$$
R_{t}^{I_{i}}=f_{i}\left(r_{t}^{0}, r_{t}^{1}, \ldots, r_{t}^{99}\right)+\varepsilon_{i}
$$
We assume the functional relation $f_i$ to be linear. We can use various strategies to design our predictive model, among many there are a few like *polynomial regression* and *inverse matrix method*. We have used the later one for our purposes.

```python
# Load the index return data
index_returns = (index_prices.pct_change() * 10000).dropna()
predictive_correlations = []
all_preds = []

for stock_return in grouped_stocks:
    # Define the matrix X
    A = np.hstack((np.ones((len(stock_return), 1)), stock_return))

    # Compute the matrix inverse of A'A
    AtA_inv = np.linalg.inv(A.T @ A)

    # Loop over each index and compute the coefficients of the linear model
    betas = []
    for i in range(len(index_returns.columns)):
        y = index_returns.iloc[:, i].values
        beta = AtA_inv @ A.T @ y
        betas.append(beta)

    # Compute the predicted index returns using the estimated coefficients
    predicted_returns = []
    for i in range(len(index_returns.columns)):
        beta = betas[i]
        predicted_return = A @ beta
        predicted_returns.append(predicted_return)

    # Convert the predicted returns to a dataframe
    predicted_returns = pd.DataFrame(np.array(predicted_returns).T, columns=index_returns.columns, index=index_returns.index)
    all_preds.append(predicted_returns)
    pred_corr = []
    for i in range(15):
        pred_corr.append(np.corrcoef(predicted_returns[str(i)],index_returns[str(i)])[0,1])
    predictive_correlations.append(pred_corr)
```

## Problem 5

To construct a trading strategy using $\mu$ and $\Sigma$ estimated in the previous part, we can use the predicted returns $\hat{R_{t}^{I_{i}}}$ and estimated covariance $\Sigma$ to optimize our portfolio weights. We can use the following steps:

For each time $t$, we calculate the predicted returns $\hat{R_{t}^{I_{i}}}$ using the polynomial regression model we built in the previous part.
We estimate the covariance matrix $\Sigma$ using the historical stock price data.
We solve for the optimal portfolio weights $w$ that maximize the expected return subject to a constraint on the maximum volatility (or risk). We can use the mean-variance optimization framework to solve for $w$, where the expected return is given by $\mu = \hat{R_{t}}$ and the risk is given by $\sqrt{w^{T}\Sigma w}$. The optimization problem can be formulated as:

1. $\mathrm{maximize}  \mu^Tw$
2. $\mathrm{subject to} \hspace{0.2em}  w^T \Sigma w \leq \sigma^2_{max}$
3. $\sum _i w_i = 0$
 $-1 \leq w_i \leq 1 , \forall i$

```python
import numpy as np

# Load mu and cov from previous part
mu = final_pred
cov = cov_matrix

# Function to get the allocation for each index at each time
def get_allocation(returns, mu, cov):
    # Calculate the predicted return for each index at each time
    index_returns = mu
    
    # Calculate the allocation using the predicted returns and covariance matrix
    inv_cov = np.linalg.inv(cov)
    allocation = np.dot(inv_cov, index_returns.T)
    
    # Constrain the allocation to be between -1 and 1
    allocation = np.clip(allocation, -1, 1)
    
    # Normalize the allocation so that the sum is 0
    allocation = allocation / np.sum(allocation)
    
    return allocation

# Get the allocation for each index at each time
allocations = []
for i in [0,4,6]:
    allocation = get_allocation(index_returns[str(i)], mu, cov)
    allocations.append(allocation)
alloc = pd.DataFrame(allocations[0])
```
