```python
import numpy as np

stocks = pd.DataFrame(stock_prices.columns)
returns = []
variance = []
for i in range(100):
    returns.append(stock_prices[str(i)].pct_change().mean() * 10**4)
    variance.append((stock_prices[str(i)].pct_change().std()) * 100) 
stocks['return'] = returns
stocks['variance'] = variance
stocks.drop([0],axis=1,inplace=True)
stocks.head()
```

```python
from sklearn.cluster import KMeans

X =  stocks.values #Converting ret_var into nummpy array
sse = []
for k in range(2,100):
    
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(X)
    
    sse.append(kmeans.inertia_) #SSE for each n_clusters
plt.plot(range(2,100), sse)
plt.title("Elbow Curve")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = []
for num_clusters in range(2,100):
 
 # initialise kmeans
 kmeans = KMeans(n_clusters=num_clusters)
 kmeans.fit(X)
 cluster_labels = kmeans.labels_
 
 # silhouette score
 silhouette_avg.append(silhouette_score(X, cluster_labels))
 
plt.plot(range(2,100),silhouette_avg,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Silhouette score') 
plt.title('Silhouette analysis For Optimal k')
plt.show()
print(silhouette_avg)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
import matplotlib.pyplot as plt

# load your data into a numpy array, assuming the shape is (n_samples, n_features)

# specify the range of cluster numbers to evaluate
k_range = range(2, 100)

# initialize an empty list to store the average silhouette scores for each value of k
silhouette_scores = []

# loop through each value of k and compute the silhouette score for each object in the dataset
for k in k_range:
    # apply KMeans clustering with k clusters and fit the data
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    
    # compute the average silhouette score for all objects in the dataset
    silhouette_avg = silhouette_score(X, labels)
    silhouette_scores.append(silhouette_avg)

# plot the average silhouette score against the number of clusters
plt.plot(k_range, silhouette_scores, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Average Silhouette Score')
plt.title('Silhouette Analysis for KMeans Clustering')
plt.show()

# select the value of k that gives the highest silhouette score as the optimal number of clusters
optimal_k = k_range[np.argmax(silhouette_scores)]
print('Optimal number of clusters:', optimal_k)
```

# Problem 3: 
We adopted the strategy of Matrix Inverse by assuming the functional relation (1) to be linear:
$$
R_{t}^{I_{i}}=f_{i}\left(r_{t}^{0}, r_{t}^{1}, \ldots, r_{t}^{99}\right)+\varepsilon_{i}
$$
We assume the functional relation $f_i$ to be linear

```python
# Loading the stock return data
stock_returns = (stock_prices.pct_change() * 10000).dropna()
stock_returns = stock_returns.loc[:, sectors[3]]
# Load the index return data
index_returns = (index_prices.pct_change() * 10000).dropna()

# Define the matrix X
A = np.hstack((np.ones((len(stock_returns), 1)), stock_returns))

# Compute the matrix inverse of A'A
AtA_inv = np.linalg.inv(A.T @ A)

# Loop over each index and compute the coefficients of the linear model
betas = []
for i in range(len(index_returns.columns)):
    y = index_returns.iloc[:, i].values
    beta = AtA_inv @ A.T @ y
    betas.append(beta)

# Compute the predicted index returns using the estimated coefficients
predicted_returns = []
for i in range(len(index_returns.columns)):
    beta = betas[i]
    predicted_return = A @ beta
    predicted_returns.append(predicted_return)

# Convert the predicted returns to a dataframe
predicted_returns = pd.DataFrame(np.array(predicted_returns).T, columns=index_returns.columns, index=index_returns.index)
```

```python
# Compute the residuals of the linear models
residuals = index_returns - predicted_returns

# Print the mean and standard deviation of the residuals
print('Mean residuals:')
print(residuals.mean())
print()
print('Standard deviation of residuals:')
print(residuals.std())
```
